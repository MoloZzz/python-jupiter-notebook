Написати програмний код, який отримує текст від користувача, аналізує його
тональність (sentiment) та виводить рівень позитивності/негативності (polarity) і
суб'єктивності (subjectivity).

```python
!pip install textblob
!pip install language-tool-python
#словники для точнішого аналізу
import nltk
nltk.download('punkt')

from textblob import TextBlob

def analyze_sentiment(text):
    blob = TextBlob(text)
    polarity = blob.sentiment.polarity      # Від -1 (негатив) до +1 (позитив)
    subjectivity = blob.sentiment.subjectivity  # Від 0 (об'єктивно) до 1 (суб'єктивно)

    print(f"Тональність (Polarity): {polarity:.2f}")
    print(f"Суб'єктивність (Subjectivity): {subjectivity:.2f}")
```

```python
user_text = input("Enter text for analyse: ")
analyze_sentiment(user_text)
```

- Polarity: від -1.0 (дуже негативний) до +1.0 (дуже позитивний).

- Subjectivity: від 0.0 (дуже об'єктивний текст) до 1.0 (дуже суб'єктивний текст).

Користувач вводить речення з граматичними помилками, реалізувати програмний
код для виправлення їх.

```python
!pip install transformers sentencepiece

from transformers import pipeline

# Завантажуємо модель
corrector = pipeline("text2text-generation", model="pszemraj/flan-t5-large-grammar-synthesis")

def correct_text(text):
    corrected = corrector(f"grammar: {text}", max_length=256)[0]["generated_text"]
    print(f"Виправлений текст: {corrected}")
```

```python
# Отримати текст від користувача
user_text = input("Введіть речення з помилками: ")

# Виправити
correct_text(user_text)
```
