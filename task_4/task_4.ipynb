$
\begin{cases}
4.2x + 10y = 136 \\
-10.1x + 13.8y = 132 \\
18.3x - 7.6y = 108
\end{cases}
$

```python
import numpy as np
import matplotlib.pyplot as plt


def line1(x): return (136 - 4.2 * x) / 10
def line2(x): return (132 + 10.1 * x) / 13.8
def line3(x): return (108 - 18.3 * x) / -7.6

# 2.   
x_vals = np.linspace(-10, 20, 400)

fig, axs = plt.subplots(1, 3, figsize=(18, 5), sharey=True)

axs[0].plot(x_vals, line1(x_vals), color='purple', linestyle='--')
axs[0].set_title(" 1: 4.2x + 10y = 136")
axs[0].grid(True)

axs[1].plot(x_vals, line2(x_vals), color='darkorange', linestyle='-.')
axs[1].set_title(" 2: -10.1x + 13.8y = 132")
axs[1].grid(True)

axs[2].plot(x_vals, line3(x_vals), color='teal', linestyle=':')
axs[2].set_title(" 3: 18.3x - 7.6y = 108")
axs[2].grid(True)

plt.tight_layout()

# 3. ' 
plt.figure(figsize=(8, 16), dpi=100)
plt.plot(x_vals, line1(x_vals), label='4.2x + 10y = 136', color='purple', linestyle='--')
plt.plot(x_vals, line2(x_vals), label='-10.1x + 13.8y = 132', color='darkorange', linestyle='-.')
plt.plot(x_vals, line3(x_vals), label='18.3x - 7.6y = 108', color='teal', linestyle=':')

plt.xlabel("x")
plt.ylabel("y")
plt.title("  ")
plt.grid(True, color='gray', linestyle='dotted')
plt.legend()

# 4.   
A1 = np.array([[4.2, 10], [-10.1, 13.8]])
b1 = np.array([136, 132])
p1 = np.linalg.solve(A1, b1)

A2 = np.array([[4.2, 10], [18.3, -7.6]])
b2 = np.array([136, 108])
p2 = np.linalg.solve(A2, b2)

A3 = np.array([[-10.1, 13.8], [18.3, -7.6]])
b3 = np.array([132, 108])
p3 = np.linalg.solve(A3, b3)

#  
plt.plot(*p1, 'o', label=' 1&2', color='black')
plt.plot(*p2, 'o', label=' 1&3', color='blue')
plt.plot(*p3, 'o', label=' 2&3', color='green')

#     
plt.text(p1[0]+0.2, p1[1], f'{p1.round(2)}', color='black')
plt.text(p2[0]+0.2, p2[1], f'{p2.round(2)}', color='blue')
plt.text(p3[0]+0.2, p3[1], f'{p3.round(2)}', color='green')

#   
x_poly = [p1[0], p2[0], p3[0]]
y_poly = [p1[1], p2[1], p3[1]]
plt.fill(x_poly, y_poly, color='lightgray', alpha=0.5)

plt.savefig("system_plot.jpg", dpi=100)
plt.savefig("system_plot.png", dpi=100)
plt.savefig("system_plot.svg", dpi=100)

plt.show()

p1, p2, p3
```
